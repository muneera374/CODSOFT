This dataset is highly unbalanced

0 -->Normal Transaction

1 -->Fradulent Transaction

```python
# Separating the data for analysis
legit=cred_card[cred_card.Class==0]
fraud=cred_card[cred_card.Class==1]
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and fradulent transactions

Number of Fradulent transactions --> 492

```python
legit_sample=legit.sample(n=492)
```

Concatenating two dataframes

```python
new_cred_card=pd.concat([legit_sample,fraud],axis=0)   #axis=0 for row concatenation,axis=1 for colun concatenation
```

```python
new_cred_card.head()
```

```python
new_cred_card.tail()
```

```python
# check the distribution of the new dataset
new_cred_card["Class"].value_counts()
```

```python
new_cred_card.groupby("Class").mean()
```

Splitting the data into features and Target

```python
x=new_cred_card.drop(columns="Class",axis=1)
y=new_cred_card["Class"]
```

```python
x
```

```python
y
```

Split the data into training and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy on training data
x_train_pred=model.predict(x_train)
training_data_acc=accuracy_score(y_train,x_train_pred)
```

```python
print("Accuracy Score of training data:",training_data_acc)
```

```python
#Accuracy on test data
x_test_pred=model.predict(x_test)
test_data_acc=accuracy_score(y_test,x_test_pred)
```

```python
print("Accuracy Score of test data:",test_data_acc)
```
